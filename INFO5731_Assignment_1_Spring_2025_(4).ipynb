```python
# Your code here
import math

def evaluate_expression(expression: str) -> float:
    """
    Evaluate a mathematical expression.
    Supports +, -, *, /, ** (exponentiation), sqrt, %, and parentheses.
    """
    try:
        # Replace 'sqrt' with 'math.sqrt' for evaluation
        expression = expression.replace("sqrt", "math.sqrt")
        # Evaluate the expression safely
        result = eval(expression, {"__builtins__": None}, {"math": math})
        return result
    except ZeroDivisionError:
        raise ValueError("Error: Division by zero is not allowed.")
    except (SyntaxError, NameError):
        raise ValueError("Error: Invalid expression.")
    except Exception as e:
        raise ValueError(f"Error: {str(e)}")

def calculator():
    """
    Command-line calculator program.
    """
    print("Welcome to the Command-Line Calculator!")
    print("Supported operations: +, -, *, /, ** (exponentiation), sqrt, % (modulus)")
    print("Use parentheses for complex expressions.")
    print("Type 'exit' to quit the calculator.")

    while True:
        # Prompt the user for input
        expression = input("\nEnter an expression (or 'exit' to quit): ").strip()

        # Exit the calculator if the user types 'exit'
        if expression.lower() == "exit":
            print("Exiting the calculator. Goodbye!")
            break

        # Evaluate the expression
        try:
            result = evaluate_expression(expression)
            print(f"Result: {result}")
        except ValueError as e:
            print(e)

# Run the calculator
if __name__ == "__main__":
    calculator()
```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os

# Constants
INVENTORY_FILE = "inventory.csv"
CATEGORIES = ["Electronics", "Clothing", "Home & Kitchen", "Books", "Toys"]
MIN_STOCK_LEVEL = 10
RESTOCK_QUANTITY = 20
TOTAL_PRODUCTS = 50

# Ensure the inventory file exists
if not os.path.exists(INVENTORY_FILE):
    with open(INVENTORY_FILE, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["ID", "Name", "Category", "Quantity", "Price"])

def load_inventory():
    """Load inventory from the CSV file."""
    try:
        with open(INVENTORY_FILE, mode="r") as file:
            reader = csv.DictReader(file)
            return list(reader)
    except Exception as e:
        print(f"Error loading inventory: {e}")
        return []

def save_inventory(inventory):
    """Save inventory to the CSV file."""
    try:
        with open(INVENTORY_FILE, mode="w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["ID", "Name", "Category", "Quantity", "Price"])
            writer.writeheader()
            writer.writerows(inventory)
    except Exception as e:
        print(f"Error saving inventory: {e}")

def add_product(name, category, quantity, price):
    """Add a new product to the inventory."""
    inventory = load_inventory()
    if len(inventory) >= TOTAL_PRODUCTS:
        print("Inventory is full. Cannot add more products.")
        return
    if category not in CATEGORIES:
        print(f"Invalid category. Choose from: {', '.join(CATEGORIES)}")
        return
    product_id = str(len(inventory) + 1)
    inventory.append({
        "ID": product_id,
        "Name": name,
        "Category": category,
        "Quantity": str(quantity),
        "Price": str(price)
    })
    save_inventory(inventory)
    print(f"Product '{name}' added successfully.")

def delete_product(product_id):
    """Delete a product from the inventory."""
    inventory = load_inventory()
    inventory = [product for product in inventory if product["ID"] != product_id]
    save_inventory(inventory)
    print(f"Product with ID {product_id} deleted successfully.")

def update_product(product_id, name=None, category=None, quantity=None, price=None):
    """Update an existing product's details."""
    inventory = load_inventory()
    for product in inventory:
        if product["ID"] == product_id:
            if name:
                product["Name"] = name
            if category:
                if category not in CATEGORIES:
                    print(f"Invalid category. Choose from: {', '.join(CATEGORIES)}")
                    return
                product["Category"] = category
            if quantity:
                product["Quantity"] = str(quantity)
            if price:
                product["Price"] = str(price)
            save_inventory(inventory)
            print(f"Product with ID {product_id} updated successfully.")
            return
    print(f"Product with ID {product_id} not found.")

def search_products(**kwargs):
    """Search for products based on criteria."""
    inventory = load_inventory()
    results = []
    for product in inventory:
        match = True
        for key, value in kwargs.items():
            if product[key] != str(value):
                match = False
                break
        if match:
            results.append(product)
    return results

def generate_report():
    """Generate a report showing item values and category breakdowns."""
    inventory = load_inventory()
    total_value = 0
    category_breakdown = {category: {"count": 0, "value": 0} for category in CATEGORIES}

    for product in inventory:
        value = int(product["Quantity"]) * float(product["Price"])
        total_value += value
        category_breakdown[product["Category"]]["count"] += 1
        category_breakdown[product["Category"]]["value"] += value

    print("\n--- Inventory Report ---")
    print(f"Total Inventory Value: ${total_value:.2f}")
    print("\nCategory Breakdown:")
    for category, data in category_breakdown.items():
        print(f"{category}: {data['count']} items, Total Value: ${data['value']:.2f}")

def automate_restocking():
    """Automatically restock products below the minimum stock level."""
    inventory = load_inventory()
    for product in inventory:
        if int(product["Quantity"]) < MIN_STOCK_LEVEL:
            product["Quantity"] = str(int(product["Quantity"]) + RESTOCK_QUANTITY)
    save_inventory(inventory)
    print("Restocking completed.")

def main():
    """Main function to manage the inventory system."""
    while True:
        print("\n--- Inventory Management System ---")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Products")
        print("5. Generate Report")
        print("6. Automate Restocking")
        print("7. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            name = input("Enter product name: ")
            category = input(f"Enter category ({', '.join(CATEGORIES)}): ")
            quantity = int(input("Enter quantity: "))
            price = float(input("Enter price: "))
            add_product(name, category, quantity, price)

        elif choice == "2":
            product_id = input("Enter product ID to delete: ")
            delete_product(product_id)

        elif choice == "3":
            product_id = input("Enter product ID to update: ")
            name = input("Enter new name (leave blank to skip): ")
            category = input(f"Enter new category ({', '.join(CATEGORIES)}, leave blank to skip): ")
            quantity = input("Enter new quantity (leave blank to skip): ")
            price = input("Enter new price (leave blank to skip): ")
            update_product(product_id, name or None, category or None, int(quantity) if quantity else None, float(price) if price else None)

        elif choice == "4":
            search_by = input("Search by (ID/Name/Category): ").lower()
            value = input(f"Enter {search_by}: ")
            results = search_products(**{search_by: value})
            if results:
                for product in results:
                    print(product)
            else:
                print("No products found.")

        elif choice == "5":
            generate_report()

        elif choice == "6":
            automate_restocking()

        elif choice == "7":
            print("Exiting the system.")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import csv
import os
import random

INVENTORY_FILE = "inventory.csv"
CATEGORIES = ["Electronics", "Clothing", "Groceries", "Furniture", "Books"]

class InventoryManager:
    def __init__(self, filename=INVENTORY_FILE):
        self.filename = filename
        self._initialize_inventory()

    def _initialize_inventory(self):
        if not os.path.exists(self.filename):
            self.generate_sample_inventory()

    def generate_sample_inventory(self):
        with open(self.filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["ID", "Name", "Category", "Quantity", "Price", "Restock Level"])

    def _read_inventory(self):
        try:
            with open(self.filename, mode='r') as file:
                reader = csv.DictReader(file)
                data = list(reader)
                if not data:
                    print("No data found.")
                return data
        except Exception as e:
            print(f"Error reading inventory file: {e}")
            return []

    def _write_inventory(self, inventory):
        try:
            with open(self.filename, mode='w', newline='') as file:
                writer = csv.DictWriter(file, fieldnames=["ID", "Name", "Category", "Quantity", "Price", "Restock Level"])
                writer.writeheader()
                writer.writerows(inventory)
        except Exception as e:
            print(f"Error writing to inventory file: {e}")

    def add_product(self, name, category, quantity, price, restock_level):
        inventory = self._read_inventory()
        product_id = max([int(item["ID"]) for item in inventory], default=0) + 1
        inventory.append({
            "ID": product_id,
            "Name": name,
            "Category": category,
            "Quantity": quantity,
            "Price": price,
            "Restock Level": restock_level
        })
        self._write_inventory(inventory)

    def delete_product(self, product_id):
        inventory = self._read_inventory()
        inventory = [item for item in inventory if int(item["ID"]) != product_id]
        self._write_inventory(inventory)

    def update_product(self, product_id, **kwargs):
        inventory = self._read_inventory()
        for item in inventory:
            if int(item["ID"]) == product_id:
                for key, value in kwargs.items():
                    if key in item:
                        item[key] = value
        self._write_inventory(inventory)

    def search_product(self, search_term):
        inventory = self._read_inventory()
        if not inventory:
            print("No data found.")
            return []
        return [item for item in inventory if search_term.lower() in item["Name"].lower()]

    def generate_report(self):
        inventory = self._read_inventory()
        if not inventory:
            print("No data found.")
            return
        total_value = sum(float(item["Price"]) * int(item["Quantity"]) for item in inventory)
        category_breakdown = {}
        for item in inventory:
            category = item["Category"]
            value = float(item["Price"]) * int(item["Quantity"])
            category_breakdown[category] = category_breakdown.get(category, 0) + value

        print(f"Total Inventory Value: ${total_value:.2f}")
        print("Category Breakdown:")
        for category, value in category_breakdown.items():
            print(f"  {category}: ${value:.2f}")

    def auto_restock(self):
        inventory = self._read_inventory()
        if not inventory:
            print("No data found.")
            return
        for item in inventory:
            if int(item["Quantity"]) < int(item["Restock Level"]):
                item["Quantity"] = int(item["Restock Level"]) * 2
        self._write_inventory(inventory)

if __name__ == "__main__":
    manager = InventoryManager()
    manager.generate_report()
    manager.auto_restock()
    manager.generate_report()
```
